```python
#1-1
#math.exp() 함수를 사용하기 위해 math 모듈 import
import math

def sigmoid(x):
    #sigmoid 수식 구현
    y = 1 / (1 + math.exp(-x))
    return y

#입력값을 실수로 받기 위해 float() 사용
n = float(input())
print(sigmoid(n))
```

```python
#1-2
import numpy as np
#math.exp() 함수를 사용하기 위해 math 모듈 import
import math

def sigmoid_list(x_list):
    #0을 요소로 가진 2차원 벡터 생성
    y = np.zeros((len(x_list), 1))
    #각 벡터에 대해 시그모이드 수식의 결과를 저장
    for i in range(len(x_list)): y[i] = 1 / (1 + math.exp(-x_list[i]))
    return y

#여러 값을 입력받기 위해 input().split() 사용, 모든 입력값을 실수형으로 받기 위해 float() 사용, 리스트로 저장하기 위해 list() 사용
n = list(map(float, input().split()))
print(sigmoid_list(n))
```

```python
#2-2
#pandas 이용을 위해 import
import pandas as pd
#함수 시각화
from matplotlib import pyplot as plt

#시각화 함수
def visualize(df, w, b):
    for idx in range(len(df)):
        #선형 함수의 시작 x좌표 지정
        x1_1 = 0
        #선형 함수의 시작 x좌표에 따른 y좌표 지정
        y1_1 = (-w[0] * x1_1 - b) / w[1]

        #선형 함수의 마지막 x좌표 지정
        x1_2 = 0.8
        #선형 함수의 마지막 x좌표에 따른 y좌표 지정
        y1_2 = (-w[0] * x1_2 - b) / w[1]

        #class값이 1인 경우
        if df.iloc[idx, 2] == 1:
            #x1을 x좌표로, x2를 y좌표로 하는 파란색 점을 찍음
            plt.plot(df.iloc[idx, 0], df.iloc[idx, 1], 'bo')
        #그렇지 않은 경우
        else:
            #x1을 x좌표로, x2를 y좌표로 하는 빨간색 점을 찍음
            plt.plot(df.iloc[idx, 0], df.iloc[idx, 1], 'ro')

        #(x1_1, y1_1)부터 (x1_2, y1_2)까지 그려진 직선을 출력
        plt.plot([x1_1, x1_2], [y1_1, y1_2])

#weights의 초기값을 각각 0.5, 0.5로 지정
w = [0.5, 0.5]
#bias의 초기값 0으로 지정
b = 0
#pandas 속 read_csv()함수를 이용하여 pandas 파일 입출력 객체 지정
df = pd.read_csv("ann_data.csv")

#데이터 시각화
visualize(df, w, b)
```

```python
#2-3
#pandas를 이용하기 위해 import
import pandas as pd
import numpy as np
#math.exp() 함수를 사용하기 위해 math 모듈 import
import math

def sigmoid_list(x_list):
    y = np.zeros((len(x_list), 1))
    #sigmoid 수식을 모든 x_list 요소에 적용, 결과값을 y에 저장
    for i in range(len(x_list)): y[i] = 1 / (1 + math.exp(-x_list[i]))
    return y

def zeta(w, b, x1, x2):
    #0을 요소로 가진 2차원 벡터 생성
    z = np.zeros((len(x1), 1))
    #벡터의 길이만큼 반복
    for idx in range(len(x1)):
        #z 벡터의 각 요소의 값을 동일한 인덱스의 x1, x2 값을 활용하여 계산한 값으로 지정
        z[idx] = w[0] * x1[idx] + w[1] * x2[idx] + b

    return z

#pandas 속 read_csv()함수를 이용하여 pandas 파일 입출력 객체 지정
df = pd.read_csv("ann_data.csv")
#ann_data.csv에서 x1 데이터 추출
x1 = np.array(df.iloc[:, 0:1])
#ann_data.csv에서 x2 데이터 추출
x2 = np.array(df.iloc[:, 1:2])

#bias의 초기값 0으로 지정
b = 0
#weights의 초기값을 각각 0.5, 0.5로 지정
w = [0.5, 0.5]

#zeta 함수의 결과값을 sigmoid_list 함수의 인자로 사용한 결과값 저장
y = sigmoid_list(zeta(w, b, x1, x2))
print(y)
```

```python
#2-4
import pandas as pd
import math
import numpy as np

def sigmoid_list(x_list):
    y = np.zeros((len(x_list), 1))
    #sigmoid 수식을 모든 x_list 요소에 적용, 결과값을 y에 저장
    #math.exp() 사용 시 DeprecationWarning 발생, np.exp()로 대체
    for i in range(len(x_list)): y[i] = 1 / (1 + np.exp(-x_list[i]))
    return y

def zeta(w, b, x1, x2):
    #0을 요소로 가진 2차원 벡터 생성
    z = np.zeros((len(x1), 1))
    #벡터의 길이만큼 반복
    for idx in range(len(x1)):
        #z 벡터의 각 요소의 값을 동일한 인덱스의 x1, x2 값을 활용하여 계산한 값으로 지정
        z[idx] = w[0] * x1[idx] + w[1] * x2[idx] + b

    return z

#pandas 속 read_csv()함수를 이용하여 pandas 파일 입출력 객체 지정
df = pd.read_csv("ann_data.csv")

#weights의 초기값을 각각 0.5, 0.5로 지정
w = [0.5, 0.5]
#bias의 초기값 0으로 지정
b = 0
#learning rate 변수 mu 초기값을 0.2로 지정
mu = 0.2
#ann_data.csv에서 x1 데이터 추출
x1 = np.array(df.iloc[:, 0:1])
#ann_data.csv에서 x2 데이터 추출
x2 = np.array(df.iloc[:, 1:2])
#ann_data.csv에서 class 데이터 추출
tn = np.array(df.iloc[:, 2:3])

#5000번 반복
for i in range(5000):
    #y를 zeta 함수를 적용한 결과값을 sigmoid_list 함수의 인자로 사용한 결과값으로 지정
    y = sigmoid_list(zeta(w, b, x1, x2))
    #학습에 따른 weights 파라미터 재설정
    w[0] += mu * np.sum((tn - y) * y * (np.ones((len(x1), 1)) - y) * x1)
    w[1] += mu * np.sum((tn - y) * y * (np.ones((len(x2), 1)) - y) * x2)
    #학습에 따른 bias 파라미터 재설정
    b += mu * np.sum((tn - y) * y * (np.ones((len(x1), 1)) - y))

#5000번의 학습으로 변화한 weights, bias 값 출력
print(w)
print(b)
```
