```python
#2-4
import pandas as pd
import numpy as np

def sigmoid_list(x_list):
    y = np.zeros((len(x_list), 1))
    for i in range(len(x_list)): y[i] = 1 / (1 + np.exp(-x_list[i]))
    return y

def zeta(w, b, x1, x2):
    z = np.zeros((len(x1), 1))
    for idx in range(len(x1)):
        z[idx] = w[0] * x1[idx] + w[1] * x2[idx] + b

    return z

df = pd.read_csv("ann_data.csv")

w = [0.5, 0.5]
b = 0
mu = 0.2
x1 = np.array(df.iloc[:, 0:1])
x2 = np.array(df.iloc[:, 1:2])
tn = np.array(df.iloc[:, 2:3])

for i in range(5000):
    y = sigmoid_list(zeta(w, b, x1, x2))
    w[0] += mu * np.sum((tn - y) * y * (np.ones((len(x1), 1)) - y) * x1)
    w[1] += mu * np.sum((tn - y) * y * (np.ones((len(x2), 1)) - y) * x2)
    b += mu * np.sum((tn - y) * y * (np.ones((len(x1), 1)) - y))
print(w)
print(b)
```
